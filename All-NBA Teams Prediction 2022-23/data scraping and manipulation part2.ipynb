```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to fetch player links for a specific year
def fetch_all_nba_links(year):
    url = f"https://www.basketball-reference.com/awards/awards_{year}.html"

    response = requests.get(url)
    content = response.content
    soup = BeautifulSoup(content, 'html.parser')

    all_nba_table = soup.find('table', {'id': 'leading-all-nba'})

    rows = all_nba_table.find_all('tr')
    player_links = []

    for row in rows:
        cols = row.find_all('td')
        for col in cols:
            link = col.find('a')
            if link:
                player_name = link.text
                player_href = link['href']
                player_links.append((year, player_name, player_href))
    return player_links

# List of years to scrape
years = [2019, 2020, 2021, 2022]

# Initialize an empty list to store the player data
player_data = []

# Fetch the player links for each year
for year in years:
    player_data.extend(fetch_all_nba_links(year))

# Create a DataFrame with the player data
player_df = pd.DataFrame(player_data, columns=['Year', 'Player', 'Link'])

# Print the DataFrame
print(player_df)
```
