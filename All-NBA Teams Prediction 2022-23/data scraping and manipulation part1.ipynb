```python
player_name = df.Player.unique()
```

```python
import pandas as pd
from nba_api.stats.static import players

# Get all NBA players
nba_players = players.get_players()

# List of desired player names
desired_players = player_name


# Filter the players and get their IDs
filtered_players = [
    {"id": player["id"], "name": player["full_name"]}
    for player in nba_players
    if player["full_name"] in desired_players
]

# Create a DataFrame with the filtered players' data
players_df = pd.DataFrame(filtered_players)

# Display the DataFrame
print(players_df)
```

```python
ids = players_df.id.unique()
```

```python
df = pd.merge(df,players_df,left_on='Player',right_on='name',how='inner')
```

```python
df.drop(columns='name',inplace=True)
```

```python
df['season'] = df['year'].apply(lambda x: f"{x-1}-{str(x)[-2:]}")
```

```python
df=df.sort_values(by = ['season', 'Share'], ascending = [True, False])
```
